```python
def generate_points_around_line(a, b, c, n, x_values):
    
    noise = np.random.normal(0, 3, n)
    y_values = a * (x_values**2) + b * x_values + c + noise
    return x_values, y_values
```

```python
num_samples, num_features = 10, 5
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)
data = np.random.rand(num_samples, num_features)
```

```python
def standardize(x):
    """Stadartize the input data x

    Args:
        x: numpy array of shape=(num_samples, num_features)

    Returns:
        standartized data, shape=(num_samples, num_features)

    """

    centered_data = x - np.mean(x, axis=0)
    std_data = centered_data / np.std(centered_data, axis=0)

    return std_data
```

```python
std_data = standardize(data)
```

```python
print(std_data, "\n\n", np.mean(std_data, axis=0), "\n\n", np.std(std_data, axis=0))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_points_around_line(a, b, n):
    x_values = np.linspace(-10, 10, n)
    noise = np.random.normal(0, 1, n)
    y_values = a * x_values + b + noise
    return x_values, y_values
```

```python
def compute_elementwise_error(a, b, x_values, y_values):
    y_hat = a * x_values + b
    return y_values - y_hat
```

```python
def compute_average_error(errors):
    return np.mean(errors**2)
```

```python
# Parameters
n = 100
interval = np.arange(-10, 10.1, 0.1)
```

```python
# Generate grid of a and b values
a_values, b_values = np.meshgrid(interval, interval) # 2d array to get values of a and b but in form that 2 arrays of a and b also there
```

The combination of a_values and b_values creates a grid of coordinates, where a_values contains the x-coordinates and b_values contains the y-coordinates of the grid points. This is typically used in scenarios where you need to evaluate functions or perform operations that depend on two variables across a defined range. For example, if you were plotting a function z = f(a, b) over a range of a and b values, these arrays would allow you to compute the z-values for a whole grid of (a, b) points at once.

a_values is array of a repeated over the entire interval in each row, and b_values is column of b repeated over the entire interval 


```python
a_values, b_values
```

```python
# Initialize an array to store average errors for each combination of a and b
average_errors = np.zeros_like(a_values)
```

```python
# Loop through all combinations of a and b
for i in range(len(interval)):
    for j in range(len(interval)):
        a = a_values[i, j]
        b = b_values[i, j]

        # Generate noisy points around the line y = ax + b
        x_values, y_values = generate_points_around_line(a, b, n)

        # Compute element-wise errors
        errors = compute_elementwise_error(a, b, x_values, y_values)

        # Compute average error for this combination of a and b
        average_errors[i, j] = compute_average_error(errors)
```

```python
# Plot the error surface
fig = plt.figure(figsize=(16, 14))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(a_values, b_values, average_errors, cmap='viridis')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('Average Error (E)')
ax.set_title('Average Error Surface')
plt.show()
```
