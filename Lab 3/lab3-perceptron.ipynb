# Write the Perceptron algorithm as discussed in class

```python
class Perceptron:

    def __init__(self):
        
        self.weights = None
    
    def train(self, x_train, y_train):
        
        samples, n_features = x_train.shape
        
        self.weights = np.zeros(n_features + 1) # 1 for bias
        prev_weight = self.weights
        
        #print(prev_weight)
        binary_y = np.array([1 if i > 0 else -1 for i in y_train])
        
        #print(n_features, samples)
        x_with_bias = np.c_[x_train, np.ones(samples)]
        #print(x_with_bias)
        

        while True:
            misclassifications = 0
            for i in range(samples):
                # updating the weights in the case of misclassification
                if (y_train[i] * np.dot(prev_weight, x_with_bias[i]) <= 0):
                    self.weights = prev_weight + y_train[i] * x_with_bias[i]
                    prev_weight = self.weights
                    misclassifications += 1
                    
            
            if (misclassifications == 0):
                break
            #raise NotImplementedError
                
    def test(self, x_test):
        x_test_with_bias = np.c_[x_test, np.ones(shape = len(x_test), dtype = int)]
        
        y_predicted = []
        final_weight = self.weights
        
        for x in x_test_with_bias:
            if (np.dot(final_weight, x) > 0):
                y_predicted.append(1)
            else:
                y_predicted.append(-1)
                
        return np.array(y_predicted)
        #raise NotImplementedError
```

```python
def compute_accuracy(y_test, y_predict):
    n = len(y_test)
    classified = 0
    for i in range(n):
        if (y_test[i] == y_predict[i]):
            classified += 1
    
    accuracy = (classified / n) * 100
    return accuracy
    #raise NotImplementedError
```

```python
x_train = np.array([[3], [4], [6], [7], [8], [9], [5], [6], [9.5]])
y_train = np.array([-1, -1, -1, 1, 1, 1, -1, -1, 1])
```

```python
print(x_train)
x_train.shape # 9 examples and 1 represents the number of features
```

```python
model = Perceptron()
model.train(x_train, y_train)
```

```python
x_test = np.array([[6.8], [7.2], [8.5], [11.4]])
y_test = [1] * len(x_test)
```

```python
y_pred = model.test(x_test)
y_pred
```

```python
accuracy = compute_accuracy(y_test, y_pred)
print(f"The accuracy of the perceptron algorithm model is: {accuracy} %.")
```
