## Step 2: SVM Implementation from Scratch (Binary Classification)
You will now implement the Linear SVM for binary classification using the hinge loss and gradient descent for DrugY versus the rest.

Tasks:

1. Initialize weights, bias, and learning rate.
2. Implement the gradient descent update rule for the SVM.
3. Train the SVM using the hinge loss function.
4. Evaluate the model on the test set.

```python
# hyperplane = wTx+b = 0
```

```python
import numpy as np

# SVM Parameters

class LinearSVM:
    def __init__(self, num_iter=100, C = 1.0, learning_rate = 0.01):
        self.num_iter = num_iter
        self.alpha = learning_rate
        self.C = C
        self.weights = None
        self.bias = None
        


    # Initialize weights and bias
    def fit(self, X, y):
        print("self.C", self.C)
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0


        for _ in range(self.num_iter):
            # Gradient Descent
            dw, db = self.gradient(X,y)
            self.weights = self.weights- self.alpha * dw
            self.bias =self.bias- self.alpha * db


    # Hinge Loss function
    def hinge_loss(self, X, y):
        term = 1-y*(np.dot(X, self.weights)+self.bias)
        slack = np.maximum(0, term)
        return self.C *np.sum(slack) #hinge loss
        #return NotImplementedError

    # Gradient Descent Update
    def gradient(self, X, y):
        #yi*(Wâ‹…xi+b)<1 checks if the data point xi is either misclassified or within the margin. 
        # If this condition is true, it contributes to the hinge loss. If it's false, the data point is correctly classified and outside the margin, contributing no loss (assuming no regularization).
        #regularization loss differentiation is added to the gradient
        dw = np.zeros(len(self.weights))
        db = 0
        for i in range(len(y)):
            if y[i]*(np.dot(X[i], self.weights)+self.bias) < 1:
                dw = dw-y[i]*X[i]
                db = db-y[i]
        print(self.C)
        C = self.C
        dw = dw + (1/(2*C))*self.weights
        #averaging the gradients
        dw = dw/len(y)
        db = db/len(y)
        return dw, db

    # Prediction Function
    def predict(self, X):
        y_pred = np.dot(X,self.weights)+self.bias
        return [1 if y_pred[i]>=0 else -1 for i in range(len(y_pred))]


    # Evaluate on Test Set
    def evaluate(self, X, y):
        y_pred = self.predict(X)
        accuracy = np.sum(y_pred == y)/len(y)
        return accuracy
```

## Step 3: Impact of Slack Variable
Now, you will explore the effect of the slack variable
C on the performance of your SVM model. Try various values of
C and observe how it impacts the decision boundary and classification performance.

```python
# Different values for C
C_values = [0.01, 0.1, 1.0, 10.0, 100.0]
accuracies = []

for C in C_values:
    # Train the SVM model with the current value of C
    svm = LinearSVM(learning_rate=0.001,C=C)
    svm.fit(x_train.values, y_train.values)

    # Evaluate the model
    accuracy = svm.evaluate(x_test.values, y_test.values)
    accuracies.append(accuracy)
    print(f"C: {C}, Accuracy: {accuracy * 100:.2f}%")


```
